```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()  # устанавливаем seaborn по умолчанию для отрисовки графиков

df = pd.read_csv("caesarian.csv",delimiter=';',)
print(df.head())
print(df.info())

print(df["Age"])
```

```python
#age
sns.countplot(x=df["Age"],);
plt.suptitle("Распределение возрастов группы")
plt.show()
```

```python
#delivery_number
sns.countplot(x=df["Delivery number"],);
plt.suptitle("Распределение порядкового номера родов")
plt.show()
```

```python
#delivery_time
sns.countplot(x=df["Delivery time"],);
plt.suptitle("Распределение времени родов\n (0 = своевременные, 1 = преждевременные, 2 = опоздавшие)")
plt.show()
```

```python
#blood_pressure
sns.countplot(x=df["Blood of Pressure"],);
plt.suptitle("Распределение давления\n (0 = низкое, 1 = нормальное, 2 = высокое)")
plt.show()
```

```python
#Heart Problem
sns.countplot(x=df["Heart Problem"],);
plt.suptitle("Проблемы с сердцем\n (0 = подходящее, 1 = неподходящее)")
plt.show()
```

```python
#Caesarian
sns.countplot(x=df["Caesarian"],);
plt.suptitle("Было ли кесарево\n (0 = нет, 1 = да")
plt.show()
```

### Зависимости

```python
#Ceasarian + age
sns.countplot(data=df, x="Age", hue='Caesarian')
plt.suptitle("Зависимость возраста и кесарева")
plt.show()
```

```python
#Ceasarian + Heart Problem
sns.kdeplot(data=df, x="Heart Problem", hue='Caesarian')
plt.suptitle("Зависимость проблем с сердцем и кесарева")
plt.show()
```

```python
#Ceasarian + Delivery number
sns.kdeplot(data=df, x="Delivery number", hue='Caesarian')
plt.suptitle("Зависимость количества родов и кесарева")
plt.show()
```

```python
#Ceasarian + Delivery time
sns.kdeplot(data=df, x="Delivery time", hue='Caesarian')
plt.suptitle("Зависимость своевременности родов и кесарева")
plt.show()
```

```python
#Ceasarian + Blood of Pressure
sns.countplot(data=df, x="Blood of Pressure", hue='Caesarian')
plt.suptitle("Зависимость давления и кесарева\n0 = низкое, 1 = нормальное, 2 = высокое")
plt.show()
```
